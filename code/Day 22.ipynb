```python
import re
import itertools as it
```

```python
def parse_input(str_in):
    
    m=re.match('Player 1:([\d\s]+)+Player 2:([\d\s]+)',
              str_in,
              flags=re.MULTILINE)
    
    return ([int(x) for x in m.group(1).split()],
            [int(x) for x in m.group(2).split()])
    
```

```python
def find_winning_hand(list_1, list_2):
    while (list_1 and list_2):
        c1=list_1.pop(0)
        c2=list_2.pop(0)
        if c1>c2:
            list_1.extend([c1, c2])
        elif c2>c1:
            list_2.extend([c2, c1])
        
    return list_1 if list_1 else list_2
```

```python
def score(list_in):
    return sum(v*m for (v, m) in zip(list_in[::-1], it.count(1)))
```

```python
def day21_part1(file_in):
    
    return score(find_winning_hand(*parse_input(open(file_in).read())))
```
