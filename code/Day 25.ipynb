## Part 1

```python
def find_loop_size(key_i):

    code=1
    loop=0
    
    while code != key_i:
        
        loop+=1
        code=(code*7)%20201227
        
    return loop
```

```python
def encryption_key(key_i, loop_i):
    
    out=1
    for i in range(loop_i):
        out=(out*key_i)%20201227
        
    return out
```

```python
door_key_test=5764801
card_key_test=17807724
```

```python
assert encryption_key(door_key_test, find_loop_size(card_key_test)) == encryption_key(card_key_test, find_loop_size(door_key_test)) 
```

```python
def day25_part1(door_key, card_key):
    
    return encryption_key(door_key, 
                          find_loop_size(card_key))
```
